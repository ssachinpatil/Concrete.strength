### Dividing data

```python
x=df.drop('Strength',axis=1)
y=df['Strength']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.20)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

### eavalution on tesy data

```python
pred=model.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,pred)
print('mae',mae)
print("*"*78)
r2=r2_score(y_test,pred)
print('r2',r2)
print("*"*78)
mse=mean_squared_error(y_test,pred)
print('mse',mse)
print("*"*67)
rmse=np.sqrt(mse)
print('rmse',rmse)
print("*"*78)
addr2=1-(((1-r2)*(x_test.shape[0])-1))/(x_test.shape[0]-x_test.shape[1]-1)
addr2
```

```python
# training dataevaluion
predict=model.predict(x_train)
mae=mean_absolute_error(y_train,predict)
print('mae',mae)
print("*"*78)
r2=r2_score(y_train,predict)
print('r2',r2)
print("*"*78)
mse=mean_squared_error(y_train,predict)
print('mse',mse)
print("*"*67)
rmse=np.sqrt(mse)
print('rmse',rmse)
print("*"*78)
addr2=1-(((1-r2)*(x_train.shape[0])-1))/(x_train.shape[0]-x_train.shape[1]-1)
addr2
```
